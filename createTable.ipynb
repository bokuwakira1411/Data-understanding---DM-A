```python
df_priceHis = pd.read_excel(Path+"/FPT_PriceHistory_after_fill.xlsx")
df_priceHis['NGÀY'] = pd.to_datetime(df_priceHis['NGÀY'], format='%d/%m/%Y')
df_priceHis.head()
```

```python
# Lấy các chỉ số từ Báo cáo chỉ số tài chính của công ty
new_table = df_finIn.loc[['Biên lợi nhuận gộp', 'Biên lợi nhuận ròng', 'P/E', 'EPS (VNĐ/CP)',
                          'Tăng trưởng EPS', 'ROE LTM', 'Nợ phải trả / Vốn chủ sở hữu',
                          'Khả năng thanh toán tổng quát','Vòng quay tài sản (vòng)', 'Giá trị sổ sách (VNĐ/CP)']]
new_table = new_table.iloc[:,:16]
new_table
```

```python
df = df_priceHis[['NGÀY', 'GIÁ ĐÓNG CỬA', 'THAY ĐỔI GIÁ', '% THAY ĐỔI']].copy()
df['QUÝ'] = df['NGÀY'].dt.quarter
df['NĂM'] = df['NGÀY'].dt.year
df.head()
```

```python
last_days = df.groupby(['NĂM', 'QUÝ'])['NGÀY'].max().reset_index()
#last_days = last_days.sort_values(by='NĂM', ascending=False).reset_index(drop=True)
last_days = last_days.sort_values(by=['NĂM', 'QUÝ'], ascending=[False, False]).reset_index(drop=True)
last_days
```

```python
# Tính trung bình theo từng quý
average_values = []

for _, row in last_days.iterrows():
    last_day = row['NGÀY']

    # Chọn khoảng 14 ngày trước và sau
    mask = (df['NGÀY'] >= last_day - pd.Timedelta(days=14)) & (df['NGÀY'] <= last_day + pd.Timedelta(days=14))
    df_window = df.loc[mask]

    # Tính trung bình
    avg_close = df_window['GIÁ ĐÓNG CỬA'].mean()
    avg_change = df_window['THAY ĐỔI GIÁ'].mean()
    avg_change_percent = df_window['% THAY ĐỔI'].mean()

    # Lưu kết quả
    average_values.append({
        'NĂM': row['NĂM'],
        'QUÝ': row['QUÝ'],
        'Ngày cuối quý': last_day,
        'Giá đóng cửa': avg_close,
        'Thay đổi giá': avg_change,
        '% Thay đổi': avg_change_percent
    })
    print(df_window)
```

```python
# Chuyển thành DataFrame
avg_df = pd.DataFrame(average_values)
avg_df['QUÝ/NĂM'] = 'Q' + avg_df['QUÝ'].astype(str) + '/' + avg_df['NĂM'].astype(str)
avg_df
```

```python
# xoá Q1/2025 vì chưa hết Q1/2025
avg_df = avg_df.drop(avg_df.index[0])
avg_df = avg_df[['Giá đóng cửa', 'Thay đổi giá',	'% Thay đổi', 'QUÝ/NĂM']].transpose()
avg_df.columns = avg_df.loc['QUÝ/NĂM']
avg_df.drop(avg_df.index[-1], inplace=True)
avg_df
```
